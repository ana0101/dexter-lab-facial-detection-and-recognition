```python
yolo_facial_detection = YOLO('models/yolo_detection.pt')

detections, scores, file_names = [], [], []

for image_path in glob.glob(test_images_path):
	result = yolo_facial_detection.predict(image_path)
	for detection in result:
		for box in detection.boxes:
			x_min, y_min, x_max, y_max = box.xyxy[0].tolist()
			score = box.conf[0].item()
			object_class = box.cls[0].item()
			detections.append([int(x_min), int(y_min), int(x_max), int(y_max)])
			scores.append(score)
			file_names.append(ntpath.basename(image_path))
```

```python
np.save('evaluare/fisiere_solutie/332_Hodivoianu_Anamaria_bonus/task1/detections_all_faces.npy', np.array(detections))
np.save('evaluare/fisiere_solutie/332_Hodivoianu_Anamaria_bonus/task1/scores_all_faces.npy', np.array(scores))
np.save('evaluare/fisiere_solutie/332_Hodivoianu_Anamaria_bonus/task1/file_names_all_faces.npy', np.array(file_names))
```

```python
yolo_facial_recognition = YOLO('models/yolo_recognition.pt')

detections_dad, scores_dad, file_names_dad = [], [], []
detections_deedee, scores_deedee, file_names_deedee = [], [], []
detections_dexter, scores_dexter, file_names_dexter = [], [], []
detections_mom, scores_mom, file_names_mom = [], [], []

for image_path in glob.glob(test_images_path):
	result = yolo_facial_recognition.predict(image_path)
	for detection in result:
		for box in detection.boxes:
			x_min, y_min, x_max, y_max = box.xyxy[0].tolist()
			score = box.conf[0].item()
			object_class = box.cls[0].item()
			if object_class == 0:
				detections_dad.append([int(x_min), int(y_min), int(x_max), int(y_max)])
				scores_dad.append(score)
				file_names_dad.append(ntpath.basename(image_path))
			elif object_class == 1:
				detections_deedee.append([int(x_min), int(y_min), int(x_max), int(y_max)])
				scores_deedee.append(score)
				file_names_deedee.append(ntpath.basename(image_path))
			elif object_class == 2:
				detections_dexter.append([int(x_min), int(y_min), int(x_max), int(y_max)])
				scores_dexter.append(score)
				file_names_dexter.append(ntpath.basename(image_path))
			elif object_class == 3:
				detections_mom.append([int(x_min), int(y_min), int(x_max), int(y_max)])
				scores_mom.append(score)
				file_names_mom.append(ntpath.basename(image_path))
```
